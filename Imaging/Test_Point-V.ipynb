```python
def DCM(a,b):
    """
        Create the DCM that maps a->b, i.e., BN
        i.e., inertial (n) -> body  (b)
    """

    a = a / np.linalg.norm(a)
    b = b / np.linalg.norm(b)
    v = np.cross(a,b)
    #v /= np.linalg.norm(v)
    s = np.linalg.norm(v)
    I = np.identity(3)
    c = np.dot(a,b)
    if s < 1e-8:
        if c > 0:
            return I
        else:
            idx = np.argmin(a)
            v = np.cross(a, I[:,idx])
    V = skew(v)
    
    R = I + V + V.dot(V) * (1-c)/s**2
    
    return R

def skew(v):
    if len(v) == 4: v = v[:3]/v[3]
    skv = np.roll(np.roll(np.diag(v.flatten()), 1, 1), -1, 0)
    return skv - skv.T


print(DCM(a,b))
```
